```python
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import json
from tqdm import tqdm

socials = {}
driver = wd.Chrome()

laptop_id = 4
input_file = f'persons_part_{laptop_id}.json'
with open(input_file, 'r', encoding='utf-8') as f:
    persons_part = json.load(f)

print(f'Ноутбук {laptop_id}: Загружено {len(persons_part)} записей')

for person in tqdm(persons_part):
    name = persons_part[person]['name']
    person_url = persons_part[person]['link']
    driver.get(person_url)
    time.sleep(0.5)
    links = {}
    
    try:
        social_section = driver.find_element(By.CLASS_NAME, 'socialList')
        social_anchors = social_section.find_elements(By.TAG_NAME, 'a')
        for anchor in social_anchors:
            href = anchor.get_attribute('href')
            if href:
                if 'twitter.com' in href or 'x.com' in href:
                    links['twitter'] = href
                elif 'instagram.com' in href:
                    links['instagram'] = href
                elif 'onlyfans.com' in href:
                    links['onlyfans'] = href
                elif 'tiktok.com' in href:
                    links['tiktok'] = href
                elif 'facebook.com' in href:
                    links['facebook'] = href
    except NoSuchElementException:
        print(f"No social section found for {name}")
    
    socials[name] = links

driver.quit()


output_file = f'socials_laptop{laptop_id}.json'
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(socials, f, ensure_ascii=False, indent=4)

print(f'Ноутбук {laptop_id}: Словарь сохранён в {output_file}')
print(f'Количество записей: {len(socials_laptopi)}')
```
