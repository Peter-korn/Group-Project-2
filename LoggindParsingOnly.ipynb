```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from tqdm import tqdm
import logging
from logging.handlers import RotatingFileHandler
import configparser

cfg = configparser.ConfigParser(interpolation=None)
cfg.read("config.ini", encoding="utf-8")

lg_enabled = cfg.getboolean("logging", "enabled", fallback=True)
lg_level   = cfg.get("logging", "level", fallback="INFO").upper()
lg_file    = cfg.get("logging", "file", fallback="ParsingOnly2.log")
lg_format  = cfg.get("logging", "format", fallback="%(asctime)s %(levelname)s %(name)s %(message)s")
lg_bytes   = cfg.getint("logging", "max_bytes", fallback=5_000_000)
lg_backups = cfg.getint("logging", "backup_count", fallback=10)
lg_mode    = cfg.get("logging", "mode", fallback="w")

LOGGER_NAME = "ParsingOnly2"
py_logger = logging.getLogger(LOGGER_NAME)

py_logger.propagate = False
py_logger.handlers.clear()

if not lg_enabled:
    logging.disable(logging.CRITICAL)
else:
    py_logger.setLevel(getattr(logging, lg_level, logging.INFO))
    handler = RotatingFileHandler(
        filename=lg_file,
        mode=lg_mode,      
        maxBytes=lg_bytes,
        backupCount=lg_backups,
        encoding="utf-8"
    )
    handler.setFormatter(logging.Formatter(lg_format))
    py_logger.addHandler(handler) 

options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome(options=options)

models = []

for page in tqdm(range(1, 201)):
    url = f"https://onlymodels.com/anal-onlyfans?page={page}"
    py_logger.info(f"Открыта страница {page}: {url}")

    driver.get(url)

    try:
        cards = WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".model-card-link"))
        )
        py_logger.info(f"Страница {page}: найдено карточек: {len(cards)}")
    except Exception:
        py_logger.warning(f"На странице {page} не найдено карточек — пропускается.")
        continue


    # for card in cards:
    #     try:
    #         name = card.find_element(By.CSS_SELECTOR, ".card-title").text.strip()
    #     except:
    #         name = None
    #         py_logger.warning(f"[p{page}] Не удалось извлечь name")
    i = 0
    for card in cards:
        i += 1
        try:
            name = card.find_element(By.CSS_SELECTOR, ".card-title").text.strip()
        except Exception as e:
            name = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь name")
        

        try:
            nickname = card.find_element(By.CSS_SELECTOR, ".text-white-50.me-2").text.strip()
        except:
            nickname = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь nickname")

        try:
            price = card.find_element(By.CSS_SELECTOR, ".badge.bg-light.text-dark").text.strip()
        except:
            price = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь price")

        try:
            likes = card.find_element(By.CSS_SELECTOR, ".bi-heart-fill").find_element(By.XPATH, "../..").text.strip()
        except:
            likes = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь likes")

        try:
            description = card.find_element(By.CSS_SELECTOR, ".description-short").text.strip()
        except:
            description = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь description")

        try:
            photo = card.find_element(By.TAG_NAME, "img").get_attribute("src")
        except:
            photo = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь photo")

        try:
            link = card.get_attribute("href")
        except:
            link = None
            py_logger.warning(f"[p{page}] card {i} Не удалось извлечь link")

        models.append({
            "name": name,
            "nickname": nickname,
            "price": price,
            "likes": likes,
            "description": description,
            "photo": photo,
            "link": link,
        })

    print(f" Страница {page}: найдено {len(cards)} моделей, всего собрано {len(models)}")
    time.sleep(1.5)

driver.quit()

print(f" Всего собрано моделей: {len(models)}")
py_logger.info(f"Всего собрано моделей: {len(models)}")


```
