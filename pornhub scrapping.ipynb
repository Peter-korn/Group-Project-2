```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
persons = {}
# 1694
for c in range(1, 201):
    url = f'https://rt.pornhub.org/pornstars?page={c}'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    pornstars = soup.find_all('li', {'class':'pornstarLi performerCard'})
    models = soup.find_all('li', {'class':'modelLi performerCard'})

    for i in range(len(pornstars)):
        rank = int(pornstars[i].find('span', {'class':'rank_number'}).getText().strip())
        name = pornstars[i].find('span', {'class': 'pornStarName performerCardName'}).getText().strip()
        videos_count = pornstars[i].find('span', {'class': 'videosNumber performerCount'}).getText().strip()
        views_count = pornstars[i].find('span', {'class': 'viewsNumber performerCount'}).getText().strip()
        link = 'https://rt.pornhub.org/pornstar/' + name.lower().replace(' ', '-')
        persons[name] ={'name': name, 'rank': rank, 'link': link, 'type': 'pornstar', 'videos_count': videos_count, 'views_count': views_count}
    
    for i in range(len(models)):
        rank = int(models[i].find('span', {'class':'rank_number'}).getText().strip())
        name = models[i].find('span', {'class': 'modelName performerCardName'}).getText().strip()
        videos_count = models[i].find('span', {'class': 'videosNumber performerCount'}).getText().strip()
        views_count = models[i].find('span', {'class': 'viewsNumber performerCount'}).getText().strip()
        link = 'https://rt.pornhub.org/model/' + name.lower().replace(' ', '-')
        persons[name] ={'name': name, 'rank': rank, 'link': link, 'type': 'model', 'videos_count': videos_count, 'views_count': views_count}
```

```python
def split_dict(dictionary, num_parts):
    keys = list(dictionary.keys())
    total = len(keys)
    chunk_size = total // num_parts + (1 if total % num_parts else 0)
    parts = []
    for i in range(0, total, chunk_size):
        part = {k: dictionary[k] for k in keys[i:i + chunk_size]}
        parts.append(part)
    return parts

num_laptops = 4
parts = split_dict(persons, num_laptops)

for i, part in enumerate(parts):
    filename = f'persons_part_{i+1}.json'
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(part, f, ensure_ascii=False, indent=4)
    print(f'Сохранено: {filename} — {len(part)} записей')
```
